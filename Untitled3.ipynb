```python
import matplotlib.pyplot as plt

monthly_cols = [col for col in ecom_data.columns if 'sales_month' in col]
ecom_data['total_sales'] = ecom_data[monthly_cols].sum(axis=1)

plt.hist(ecom_data['total_sales'], bins=30, color='skyblue', edgecolor='black')
plt.title('Total Sales Distribution')
plt.show()
```

```python
print("\nCategory Distribution:")
print(ecom_data['category'].value_counts())

plt.figure(figsize=(10,5))
ecom_data['category'].value_counts().plot(kind='bar', color='teal')
plt.title('Products by Category')
plt.xticks(rotation=45)
plt.show()
```

```python
monthly_totals = ecom_data[monthly_cols].sum()
plt.plot(monthly_totals, marker='o')
plt.title('Monthly Sales Trend')
plt.xticks(range(12), [f'Month {i+1}' for i in range(12)], rotation=45)
plt.show()
```

```python
# Example: Get sales of product_id=101 across months
product_sales = ecom_data.loc[ecom_data['product_id'] == 101, monthly_cols].T
product_sales.index = range(1, 13)  # Months 1-12
```

```python
from statsmodels.tsa.arima.model import ARIMA

try:
    model = ARIMA(product_sales, order=(1, 0, 0))  # AR(1) model
    model_fit = model.fit()
    print(model_fit.summary())
    forecast = model_fit.forecast(steps=3)  # Forecast next 3 months
    print("ARIMA Forecast:", forecast)
except Exception as e:
    print(f"ARIMA failed: {e}. Possible issues:")
    print("- Data too short (need ≥20 observations)")
    print("- Non-stationary (try differencing: order=(1,1,0))")
```

```python
# Prepare data with datetime index
dates = pd.date_range(start='2023-01-01', periods=len(monthly_cols), freq='M')
monthly_totals.index = dates

# Fit ARIMA
model = ARIMA(monthly_totals, order=(1,0,0), freq='M')
model_fit = model.fit()
print(model_fit.summary())
```

```python
# A. Residual Analysis (Critical!)
model_fit.plot_diagnostics(figsize=(10,8))
plt.show()

# B. Ljung-Box Test for residual autocorrelation
from statsmodels.stats.diagnostic import acorr_ljungbox
residuals = model_fit.resid
lb_test = acorr_ljungbox(residuals, lags=[10])  # Test first 10 lags
print("Ljung-Box p-value:", lb_test.lb_pvalue)  # p > 0.05 is good
```

```python
forecast = model_fit.forecast(steps=3)
print("Forecast:\n", forecast)

# Plot with confidence intervals
forecast = model_fit.get_forecast(steps=3)
conf_int = forecast.conf_int()
plt.plot(monthly_totals, label='Actual')
forecast.predicted_mean.plot(label='Forecast')
plt.fill_between(conf_int.index, 
                 conf_int.iloc[:,0], 
                 conf_int.iloc[:,1], 
                 color='pink', alpha=0.3)
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error

# Walk-forward validation
train_size = int(len(monthly_totals) * 0.7)
train, test = monthly_totals[:train_size], monthly_totals[train_size:]
history = list(train)
predictions = []

for t in range(len(test)):
    model = ARIMA(history, order=(1,0,0))
    model_fit = model.fit()
    yhat = model_fit.forecast()[0]
    predictions.append(yhat)
    history.append(test[t])

# Calculate MAE
mae = mean_absolute_error(test, predictions)
print(f"MAE: {mae:.2f}")
```

```python
from sklearn.metrics import mean_absolute_percentage_error

print(f"MAPE: {mean_absolute_percentage_error(test, predictions)*100:.2f}%")
# MAPE < 10% is generally good for business forecasts
```
